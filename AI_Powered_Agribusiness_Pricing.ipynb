```python
import numpy as np
import pandas as pd

# -----------------------------
# RANDOM SEED FOR REPRODUCIBILITY
# -----------------------------
np.random.seed(42)

# Assume your dataframe already exists
# df = pd.read_csv("your_dataset.csv")

# -----------------------------
# MARKET & ECONOMIC FEATURES
# -----------------------------

# Corn futures price ($/bushel)
df["corn_futures_price"] = np.random.normal(6.2, 0.7, len(df)).round(2)

# Soybean futures price ($/bushel)
df["soybean_futures_price"] = np.random.normal(13.5, 1.2, len(df)).round(2)

# Fertilizer market index
df["fertilizer_market_index"] = np.random.normal(110, 15, len(df)).round(2)

# Diesel price ($/gallon)
df["diesel_price"] = np.random.normal(4.1, 0.4, len(df)).round(2)

# Inflation rate (%)
df["inflation_rate"] = np.random.uniform(2.0, 8.0, len(df)).round(2)

# Competitor product price
df["competitor_price"] = (
    df["product_price"] * np.random.uniform(0.9, 1.1, len(df))
).round(2)

# -----------------------------
# CUSTOMER & DEMAND FEATURES
# -----------------------------

# Customer segment
customer_segments = ["Small Farm", "Commercial", "Enterprise", "Cooperative"]

df["customer_segment"] = np.random.choice(
    customer_segments,
    len(df)
)

# Farm size (hectares)
df["farm_size_hectares"] = np.random.randint(10, 5000, len(df))

# Purchase frequency per year
df["purchase_frequency"] = np.random.randint(1, 15, len(df))

# Customer lifetime value ($)
df["customer_lifetime_value"] = np.random.normal(
    25000,
    8000,
    len(df)
).round(2)

# Demand index
df["demand_index"] = np.random.uniform(0.5, 1.5, len(df)).round(2)

# Regional market share (%)
df["regional_market_share_pct"] = np.random.uniform(
    1,
    35,
    len(df)
).round(2)

# -----------------------------
# TEMPORAL FEATURES
# -----------------------------

# Convert dates to datetime
df["sowing_date"] = pd.to_datetime(df["sowing_date"])
df["harvest_date"] = pd.to_datetime(df["harvest_date"])
df['timestamp'] = pd.to_datetime(df['timestamp'])
# Extract week
df["week"] = df["timestamp"].dt.isocalendar().week

# Extract quarter
df["quarter"] = df["timestamp"].dt.quarter

# Season mapping
def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Fall"

df["season"] = df["timestamp"].dt.month.apply(get_season)

# Holiday flag
df["holiday_flag"] = np.random.choice(
    [0, 1],
    len(df),
    p=[0.9, 0.1]
)

# -----------------------------
# ENVIRONMENTAL RISK FEATURES
# -----------------------------

# Drought index
df["drought_index"] = np.random.uniform(0, 5, len(df)).round(2)

# Heat stress index
df["heat_stress_index"] = (
    (df["temperature_C"] / 40) * 100
).round(2)

# Flood risk score
df["flood_risk_score"] = np.random.uniform(
    0,
    100,
    len(df)
).round(2)

# Soil health score
df["soil_health_score"] = np.random.uniform(
    50,
    100,
    len(df)
).round(2)

# Crop disease risk
disease_risk = ["Low", "Moderate", "High"]

df["crop_disease_risk"] = np.random.choice(
    disease_risk,
    len(df),
    p=[0.6, 0.3, 0.1]
)

# Water availability index
df["water_availability_index"] = np.random.uniform(
    0,
    1,
    len(df)
).round(2)

# -----------------------------
# INVENTORY & SUPPLY CHAIN
# -----------------------------

# Inventory level
df["inventory_level"] = np.random.randint(
    100,
    5000,
    len(df)
)

# Supplier risk score
df["supplier_risk_score"] = np.random.uniform(
    0,
    100,
    len(df)
).round(2)

# Logistics cost ($)
df["logistics_cost"] = np.random.normal(
    500,
    120,
    len(df)
).round(2)

# Lead time (days)
df["lead_time_days"] = np.random.randint(
    1,
    30,
    len(df)
)

# -----------------------------
# PROFITABILITY FEATURES
# -----------------------------

# Unit production cost
df["unit_cost"] = (
    df["product_price"] * np.random.uniform(0.5, 0.85, len(df))
).round(2)

# Profit margin
df["profit_margin"] = (
    (df["product_price"] - df["unit_cost"])
    / df["product_price"]
    * 100
).round(2)

# -----------------------------
# AI PRICING RECOMMENDATION
# -----------------------------

def pricing_recommendation(row):

    if row["drought_index"] > 3 and row["demand_index"] > 1.1:
        return "Increase price by 5%"

    elif row["inventory_level"] > 4000:
        return "Offer 10% discount"

    elif row["competitor_price"] < row["product_price"]:
        return "Reduce price by 3%"

    else:
        return "Maintain current pricing"

df["pricing_recommendation"] = df.apply(
    pricing_recommendation,
    axis=1
)

# -----------------------------
# AI BUSINESS EXPLANATION
# -----------------------------

def recommendation_reason(row):

    if row["drought_index"] > 3:
        return "High drought risk increasing regional demand pressure."

    elif row["inventory_level"] > 4000:
        return "Excess inventory detected."

    elif row["competitor_price"] < row["product_price"]:
        return "Competitor pricing pressure identified."

    else:
        return "Market conditions stable."

df["pricing_recommendation_reason"] = df.apply(
    recommendation_reason,
    axis=1
)

# -----------------------------
# PREVIEW DATASET
# -----------------------------

print(df.head())

# -----------------------------
# SAVE ENHANCED DATASET
# -----------------------------

df.to_csv(
    "enhanced_agribusiness_pricing_dataset.csv",
    index=False
)

print("Enhanced dataset saved successfully.")
```

```python
hormuz_scenarios = [
    "normal",
    "moderate_disruption",
    "severe_disruption",
    "closure"
]

df["hormuz_scenario"] = np.random.choice(
    hormuz_scenarios,
    len(df),
    p=[0.65, 0.20, 0.10, 0.05]
)
```

```python
def oil_shock_multiplier(scenario):

    if scenario == "normal":
        return np.random.uniform(0.98, 1.03)

    elif scenario == "moderate_disruption":
        return np.random.uniform(1.08, 1.18)

    elif scenario == "severe_disruption":
        return np.random.uniform(1.20, 1.40)

    else:
        return np.random.uniform(1.45, 1.80)

df["oil_price_shock_multiplier"] = df[
    "hormuz_scenario"
].apply(oil_shock_multiplier)
```

```python
df["adjusted_diesel_price"] = (
    df["diesel_price"]
    * df["oil_price_shock_multiplier"]
).round(2)
```

```python
df["adjusted_fertilizer_index"] = (
    df["fertilizer_market_index"]
    * (
        1
        + (
            (df["oil_price_shock_multiplier"] - 1)
            * 0.75
        )
    )
).round(2)
```

```python
df["adjusted_logistics_cost"] = (
    df["logistics_cost"]
    * (
        1
        + (
            (df["oil_price_shock_multiplier"] - 1)
            * 0.60
        )
    )
).round(2)
```

```python
df["farm_input_inflation_pct"] = (
    (
        (
            df["adjusted_fertilizer_index"]
            - df["fertilizer_market_index"]
        )
        / df["fertilizer_market_index"]
    )
    * 100
).round(2)
```

```python
df["revenue_at_risk"] = (
    df["revenue"]
    * (
        df["farm_input_inflation_pct"] / 100
    )
).round(2)
```

```python
# ---------------------------------------------------------
# Adjusted unit production cost after geopolitical shock
# ---------------------------------------------------------

df["adjusted_unit_cost"] = (
    df["unit_cost"]
    * (
        1
        + (
            df["farm_input_inflation_pct"] / 100
        )
        * 0.55
    )
).round(2)
```

```python
df["margin_compression_pct"] = (
    (
        (
            df["adjusted_unit_cost"]
            - df["unit_cost"]
        )
        / df["unit_cost"]
    )
    * 100
).round(2)
```

```python
# ---------------------------------------------------------
# Margin compression percentage
# ---------------------------------------------------------

df["margin_compression_pct"] = (
    (
        (
            df["adjusted_unit_cost"]
            - df["unit_cost"]
        )
        / df["unit_cost"]
    )
    * 100
).round(2)
```

```python
def geopolitical_strategy(row):

    if row["hormuz_scenario"] == "closure":
        return "Increase prices aggressively"

    elif row["farm_input_inflation_pct"] > 15:
        return "Adjust pricing to protect margins"

    elif row["inventory_level"] > 4000:
        return "Offer targeted discounts"

    else:
        return "Maintain current pricing"

df["geopolitical_pricing_strategy"] = df.apply(
    geopolitical_strategy,
    axis=1
)
```

```python
def geopolitical_summary(row):

    if row["hormuz_scenario"] == "closure":

        return (
            "Severe geopolitical disruption detected. "
            "Fuel and fertilizer costs are rising sharply."
        )

    elif row["farm_input_inflation_pct"] > 10:

        return (
            "Input cost inflation increasing operational risk."
        )

    else:

        return (
            "Market conditions remain stable."
        )

df["geopolitical_summary"] = df.apply(
    geopolitical_summary,
    axis=1
)
```
