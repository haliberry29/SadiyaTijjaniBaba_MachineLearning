##Adjusting for farm input and operations cost to account for geopolitical instability


```python
def oil_shock_multiplier(scenario):

    if scenario == "normal":
        return np.random.uniform(0.98, 1.03)

    elif scenario == "moderate_disruption":
        return np.random.uniform(1.08, 1.18)

    elif scenario == "severe_disruption":
        return np.random.uniform(1.20, 1.40)

    else:
        return np.random.uniform(1.45, 1.80)

df["oil_price_shock_multiplier"] = df[
    "hormuz_scenario"
df["adjusted_fertilizer_index"] = (
].apply(oil_shock_multiplier)
```

```python
df["adjusted_diesel_price"] = (
    df["diesel_price"]
    * df["oil_price_shock_multiplier"]
).round(2) #adjust oil prices to account of hormuz closure or supply shock
```

```python
df["adjusted_fertilizer_index"] = (
    df["fertilizer_market_index"] * ( 1 + ((df["oil_price_shock_multiplier"] - 1)
            * 0.75 ))).round(2)#adjust fertilizer prices to account of hormuz closure or supply shock
```

```python
df["adjusted_logistics_cost"] = ( df["logistics_cost"] * ( 1 + ((df["oil_price_shock_multiplier"] - 1)
            * 0.60))).round(2) #adjust logistics prices to account of hormuz closure or supply shock
```

```python
df["farm_input_inflation_pct"] = (
    ((df["adjusted_fertilizer_index"] - df["fertilizer_market_index"])/ df["fertilizer_market_index"]
    )* 100).round(2) #calculate percentage inflation due to hormuz closure or supply shock
```

```python
df["revenue_at_risk"] = (
    df["revenue"] * (
        df["farm_input_inflation_pct"] / 100)).round(2) #calculate revenue at risk or variability in revenue due to external factor
```

```python
# ---------------------------------------------------------
# Adjusted unit production cost after geopolitical shock
# ---------------------------------------------------------

df["adjusted_unit_cost"] = (
    df["unit_cost"]
    * (1 + (df["farm_input_inflation_pct"] / 100)* 0.55)).round(2)
```

```python
df["margin_compression_pct"] = (
    (
        (df["adjusted_unit_cost"] - df["unit_cost"] )/ df["unit_cost"] )
    * 100
).round(2)#the percentage increase in operational costs caused by inflation, geopolitical disruption, or supply-chain pressure.
#how much profitability is being squeezed
#costs are rising faster than revenue.

#This reduces:

#profitability
#operating margin
#business sustainability
```

#simulate different geopolitical strategy

```python
def geopolitical_strategy(row):

    if row["hormuz_scenario"] == "closure":
        return "Increase prices aggressively"

    elif row["farm_input_inflation_pct"] > 15:
        return "Adjust pricing to protect margins"

    elif row["inventory_level"] > 4000:
        return "Offer targeted discounts"

    else:
        return "Maintain current pricing"

df["geopolitical_pricing_strategy"] = df.apply(
    geopolitical_strategy,
    axis=1
)
```

```python
def geopolitical_summary(row):

    if row["hormuz_scenario"] == "closure":

        return (
            "Severe geopolitical disruption detected. "
            "Fuel and fertilizer costs are rising sharply."
        )

    elif row["farm_input_inflation_pct"] > 10:

        return (
            "Input cost inflation increasing operational risk."
        )

    else:

        return (
            "Market conditions remain stable."
        )

df["geopolitical_summary"] = df.apply(
    geopolitical_summary,
    axis=1
)
```
